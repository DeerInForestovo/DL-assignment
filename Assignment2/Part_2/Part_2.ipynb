````python
from matplotlib import pyplot as plt
from cnn_train import train
from cnn_train import MAX_EPOCHS_DEFAULT
loss_history, accuracy_history = train()
````

````python
fig, ax1 = plt.subplots(figsize=(10, 6))
epochs = list(range(1, MAX_EPOCHS_DEFAULT + 1))
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.plot(epochs, accuracy_history, label='Validation accuracy')
ax2 = ax1.twinx()
ax2.set_ylabel('Loss', color='r')
ax2.tick_params(axis='y', labelcolor='r')
ax2.plot(epochs, loss_history, 'r--', label='Validation loss')
fig.legend(loc='lower left', bbox_to_anchor=(0, 0), bbox_transform=ax1.transAxes)
plt.title("Validation Accuracy and Loss of CNN on CIFAR-10")
plt.show()
````
